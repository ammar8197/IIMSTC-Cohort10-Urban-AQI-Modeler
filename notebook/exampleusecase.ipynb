**Model Experimentation**

```python
import numpy as np
import pandas as pd
import joblib
```

```python
model = joblib.load("../models/pm25_xgb_model.pkl")
```

```python
sample_input = pd.DataFrame({
    'PM10': [120],
    'SO2': [10],
    'NO2': [35],
    'CO': [1.2],
    'O3': [60],
    'TEMP': [25],
    'PRES': [1010],
    'DEWP': [10],
    'RAIN': [0],
    'WSPM': [2.5],
    'month': [3],
    'hour': [18]
})
```

**Scalar score**

```python
prediction = model.predict(sample_input)

print("Predicted PM2.5:", prediction[0])
```

**Original Dataset**

```python
raw_data = pd.read_csv("C:\\Users\\rishi\\IIMSTC-Cohort10-Urban-AQI-Modeler\\data\\raw\\Beijing Multisite air Quality data.csv")
```

```python
features = [
    'PM10', 'SO2', 'NO2', 'CO', 'O3',
    'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM',
    'month', 'hour'
]
```

```python
feature_means = raw_data[features].mean()
feature_stds  = raw_data[features].std()
```

```python
mean_pm25 = raw_data['PM2.5'].mean()
std_pm25  = raw_data['PM2.5'].std()

real_pm25 = prediction * std_pm25 + mean_pm25
```

```python
def predict_pm25(sample_input):

    # Scale input (same as training)
    scaled_input = (sample_input - feature_means) / feature_stds

    # Predict scaled PM2.5
    scaled_pred = model.predict(scaled_input)[0]

    # Convert to real PM2.5
    real_pm25 = scaled_pred * std_pm25 + mean_pm25

    # Physical constraint
    real_pm25 = max(real_pm25, 0)

    return real_pm25
```

```python
result = predict_pm25(sample_input)

print("Predicted PM2.5:", result, "µg/m³")
```
